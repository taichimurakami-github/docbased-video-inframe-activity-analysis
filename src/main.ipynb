```python
from pathlib import Path

# ↓jupiter notebookは__path__が使えないらしいのでpathlibで代用
# https://kyousukekinoshita.hatenablog.com/entry/2018/10/28/191420
video_file_path = os.path.join(Path().resolve(), ".data", "EdanMeyerVpt.mp4")

video = CV2VideoUtil()
cap = video.load_file(video_file_path)
_, result_img =  video.get_video_frame_every_sec(cap, 66)

pprint(result_img)
```

```python
image = CV2ImageUtil()
j = 0
for i in range(len(result_img) - 1):
      prev_frame_gray = image.apply_grayscale(result_img[i])
      current_frame_gray = image.apply_grayscale(result_img[i+1])

      image.show(prev_frame_gray,"prev_frame_gray")
      image.show(current_frame_gray, "current_frame_gray")


      j = j + 1

      if j > 5:
            break

```
